1. Importing Library

```python
import pandas as pd
```

2. Loading Dataset

```python
df = pd.read_csv('raw_dataset.csv', encoding='latin1')
```

3. Checking for Null Values

```python
df.isnull().sum
```

4. Removing Duplicates Values

```python
df = df.drop_duplicates()
```

5. Checking DataTypes

```python
df.dtypes
```

6. Standarize Columns Names

```python
df.columns = (df.columns
              .str.strip()
              .str.lower()
              .str.replace(" ","_")
)
```

```python
df.columns
```
