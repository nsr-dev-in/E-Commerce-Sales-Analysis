1. Import Libraries

```python
import pandas as pd
from sqlalchemy import create_engine
```

2. Read CSV

```python
df = pd.read_csv("cleaned_dataset.csv", encoding="latin1")
```

3. PostgreSQL Connection

```python
engine = create_engine(
    "postgresql://postgres:admin@localhost:5432/E_Commerce_Sales_Analysis"
)
```

4. Load into PostgreSQL

```python
df.to_sql(
    "orders",          # table name
    engine,
    if_exists="replace",
    index=False
)
```
