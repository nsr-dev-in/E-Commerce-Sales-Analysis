```python
import pandas as pd
```

```python
df = pd.read_csv("cleaned_dataset.csv")
```

```python
df.dtypes
```

```python
df["order_date"] = pd.to_datetime(df["order_date"])
```

```python
df.dtypes
```

```python
total_sales = df["sales"].sum()
```

```python
total_profit = df["profit"].sum()
```

```python
total_orders = df["order_id"].nunique()
```

```python
top_sales_product = (df.groupby("product_name")["sales"].sum().sort_values(ascending=False).head(10))
```

```python
top_profit_products = (df.groupby("product_name")["profit"].sum().sort_values(ascending=False).head(10))
```

```python
loss_products = (df.groupby("product_name")["profit"].sum())
loss_products = loss_products[loss_products < 0]
```

```python
sales_by_category = (df.groupby("category")["sales"].sum().sort_values(ascending=False))
```

```python
sales_by_region = (df.groupby("region")["sales"].sum().sort_values(ascending=False))
```

```python
df["month"] = df["order_date"].dt.to_period("M")

monthly_sales = (
    df.groupby("month")["sales"]
    .sum()
)
```

```python
monthly_profit = (
    df.groupby("month")["profit"]
    .sum()
)
```
